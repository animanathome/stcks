```python
# http://stockcharts.com/school/doku.php?id=chart_school:technical_indicators:bollinger_bands
import pandas as pd
import pandas_datareader.data as web
import datetime
import numpy as np

a = [90.7, 92.9, 92.98, 91.80, 92.66, 92.66, 92.68, 92.30, 92.77, 92.54, 92.95, 93.20, 91.07, 89.83, 89.74, 90.4, 90.74, 88.02, 88.09, 88.84, 90.78, 90.54, 91.39, 90.65]
random = pd.DataFrame({'Price':a})
random["20d SMA"] = np.round(random["Price"].rolling(window = 20).mean(), 2)

random['20-period Average'] = random['Price'].rolling(20).mean()
random['20-period Average'].values[0:19] = [random['20-period Average'].values[19] for item in range(0, 19)]
random['Deviation'] = random['Price'] - random['20-period Average']
random['Deviation Squared'] = random['Deviation']**2
random['20-period Average of Deviation Squared'] = random['Deviation Squared'].rolling(20).sum()/20
random['Standard Deviation'] = random['20-period Average of Deviation Squared'].apply(np.sqrt)
random.drop(['20-period Average', '20-period Average of Deviation Squared', 'Deviation Squared', 'Deviation', '20-period Average'], axis=1, inplace=True)

random['UB'] = random["20d SMA"] + (random['Standard Deviation'] *2)
random['LB'] = random["20d SMA"] - (random['Standard Deviation'] *2)

random
```

```python
import pandas as pd
import pandas_datareader.data as web
import datetime
import numpy as np

import matplotlib.pyplot as plt   # Import matplotlib
# This line is necessary for the plot to appear in a Jupyter notebook
%matplotlib inline
# Control the default size of figures in this Jupyter notebook
%pylab inline
pylab.rcParams['figure.figsize'] = (15, 9) 

start = datetime.datetime(2016,1,1)
end = datetime.date.today()

netflix = web.DataReader("NFLX", "yahoo", start, end)
netflix.columns
netflix['10-period Average'] = netflix['Adj Close'].rolling(10).mean()
netflix['Deviation'] = netflix['Adj Close'] - netflix['10-period Average']
netflix['Deviation Squared'] = netflix['Deviation']**2
netflix['Standard Deviation'] = netflix['Deviation Squared'].rolling(10).sum()/10**2
netflix.tail()

# 10 = the number of days over which we calculate the average
a = [53.73, 53.87, 53.85, 53.88, 54.08, 54.14, 54.50, 54.30, 54.4, 54.16]
random = pd.DataFrame({'Price':a})
random['10-period Average'] = random['Price'].rolling(10).mean()
random['10-period Average'].values[0:9] = [random['10-period Average'].values[9] for item in range(0, 9)]
random['Deviation'] = random['Price'] - random['10-period Average']
random['Deviation Squared'] = random['Deviation']**2
random['10-period Average of Deviation Squared'] = random['Deviation Squared'].rolling(10).sum()/10
random['Standard Deviation'] = random['10-period Average of Deviation Squared'].apply(np.sqrt)
random
```
